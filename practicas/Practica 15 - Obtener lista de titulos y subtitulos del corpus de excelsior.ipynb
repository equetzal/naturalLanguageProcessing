## Obtener lista de titulos y subtitulos del corpus de excelsior
**Instrucciones:**
    Del corpus de *excelsior*, obtener todos los titulos y subtitulos de los artículos.

```python
import nltk
import numpy
import math
from nltk.stem import SnowballStemmer
```

```python
def deleteTrash(txt):
	import re
	good = {'\n'}
	for i in "0123456789abcdefghijklmnopqrstuvwxyz áéíóúñü.:":
		good.add(i)
	ans = ""
	for c in txt:
		if c in good:
			ans += c
	return ans
```

```python
def isTitle(sentence):
    lower = 0
    upper = 0
    for c in sentence:
        if c.isalpha():
            if c.isupper():
                upper += 1
            else:
                lower += 1
    if lower > upper:
        return True
    return False
```

```python
def getTitles(txt):
    import re
    from bs4 import BeautifulSoup
    sentences = BeautifulSoup(txt,'lxml').find_all(['h3', 'b'])
    
    titles = list()
    for sent in sentences:
        txt = sent.get_text()
        if isTitle(txt):
            cleanText = ' '.join(deleteTrash(txt.lower()).split())
            if not cleanText.isspace():
                titles.append(cleanText)
    return titles    
```

```python
f = open("./files/e961024.htm", "r", encoding="utf8")
org_txt = f.read()
f.close()
```

```python
titles = getTitles(org_txt)
for title in titles:
    print(title)
```
